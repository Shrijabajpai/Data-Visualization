```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('filter_data.csv')
```

```python
data
```

```python
data.SMOKE=data.SMOKE.map({'yes':1,'no':0})
```

```python
data.Gender=data.Gender.map({'Female':1,'Male':0})
```

```python
data
```

```python
data.NObeyesdad.unique()
```

```python
data.info()
```

# Split independent variable and dependent variable

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
x=x.values
y=y.values
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y =  le.fit_transform(y)
```

```python
y
```

# Split train and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.4,random_state=10)
```

```python
len(y_test)
```

```python
count=0
for i in y_test:
    if i==0:
        count+=1
count        
```

#### Logistic Regression Model

Train Model

```python
from sklearn.linear_model import LogisticRegression
log_r=LogisticRegression()
log_r.fit(x_train,y_train)
```

Test Model

```python
pre=log_r.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, pre)
print(cm)
accuracy_score(y_test, pre)
```

# K- nearest Neighbor Classification Model

Train Model

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=7,metric='minkowski',p=2)
knc.fit(x_train,y_train)
```

Test Model

```python
pre_knc=knc.predict(x_test)
```

```python
cm = confusion_matrix(y_test, pre_knc)
print(cm)
accuracy_score(y_test, pre_knc)
```

# Support Vector Machine Model

Train Model

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear',random_state=0)
svm.fit(x_train,y_train)
```

Test Model

```python
pre_svm=svm.predict(x_test)
```

```python
cm = confusion_matrix(y_test, pre_svm)
print(cm)
accuracy_score(y_test, pre_svm)
```

# Kernel Support Vector Machine

Train Model

```python
from sklearn.svm import SVC
ksvm=SVC(kernel='rbf',random_state=0)
ksvm.fit(x_train,y_train)
```

Test Model

```python
pre_ksvm=ksvm.predict(x_test)
```

```python
cm = confusion_matrix(y_test, pre_ksvm)
print(cm)
accuracy_score(y_test, pre_ksvm)
```

# Naive Bayes Classifier Model

Train Model

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
```

Test Model

```python
pre_nb=nb.predict(x_test)
```

```python
cm = confusion_matrix(y_test, pre_nb)
print(cm)
accuracy_score(y_test, pre_nb)
```

# Decision Tree Classifier Model

Train Model

```python
from sklearn.tree import DecisionTreeClassifier
dtr=DecisionTreeClassifier(criterion='entropy',random_state=0)
dtr.fit(x_train,y_train)
```

Test Model

```python
pre_dtr=dtr.predict(x_test)
```

```python
cm = confusion_matrix(y_test, pre_dtr)
print(cm)
accuracy_score(y_test, pre_dtr)
```

# Random Forest Classifier Model

Train Model

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=100,criterion='entropy')
rfc.fit(x_train,y_train)
```

Test Model

```python
pre_rfc=rfc.predict(x_test)
```

```python
cm = confusion_matrix(y_test, pre_rfc)
print(cm)
accuracy_score(y_test, pre_rfc)
```
