```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('obesity_data.csv')
```

```python
plt.figure(figsize=(15, 10))

# Histogram for Age, Height, and Weight
plt.subplot(3, 2, 1)
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')

plt.subplot(3, 2, 2)
sns.histplot(df['Height'], bins=20, kde=True)
plt.title('Distribution of Height')

plt.subplot(3, 2, 3)
sns.histplot(df['Weight'], bins=20, kde=True)
plt.title('Distribution of Weight')

# Countplot for categorical attributes
plt.subplot(3, 2, 4)
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')

plt.subplot(3, 2, 5)
sns.countplot(x='family_history_with_overweight', data=df)
plt.title('Family History with Overweight')

plt.subplot(3, 2, 6)
sns.countplot(x='FAVC', data=df)
plt.title('Frequency of consuming high caloric food')

plt.tight_layout()
plt.show()
```

```python
#2. Bivariate Analysis

plt.figure(figsize=(15, 8))

# Scatter plot for Age vs Weight
plt.subplot(2, 2, 1)
sns.scatterplot(x='Age', y='Weight', data=df)
plt.title('Age vs Weight')

# Boxplot for Gender vs Weight
plt.subplot(2, 2, 2)
sns.boxplot(x='Gender', y='Weight', data=df)
plt.title('Gender vs Weight')

# Violin plot for family_history_with_overweight vs Weight
plt.subplot(2, 2, 3)
sns.violinplot(x='family_history_with_overweight', y='Weight', data=df)
plt.title('Family History with Overweight vs Weight')

plt.tight_layout()
plt.show()
```

```python
# 3. Multivariate Analysis

sns.pairplot(df[['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']])
plt.suptitle('Pairplot of Numeric Attributes', y=1.02)
plt.show()
```

```python
correlation_matrix = df[['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# 4. Categorical Plots

plt.figure(figsize=(15, 10))

# Countplot for MTRANS (Mode of Transport)
plt.subplot(3, 2, 1)
sns.countplot(x='MTRANS', data=df)
plt.title('Mode of Transport Distribution')

# Bar plot for CAEC (Consumption of food between meals)
plt.subplot(3, 2, 2)
sns.barplot(x='CAEC', y='Weight', data=df)
plt.title('Average Weight by Consumption of Food Between Meals')

# Stacked bar plot for SMOKE vs NObeyesdad
smoke_obesity = pd.crosstab(df['SMOKE'], df['NObeyesdad'])
smoke_obesity.plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot of Smoking vs Obesity Class')
plt.xlabel('Smoking Status')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# 5. Multivariate Analysis
plt.figure(figsize=(15, 10))
sns.pairplot(df[['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF']], diag_kind='kde')
plt.suptitle('Pairplot of Numeric Attributes', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = df[['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Attributes')
plt.show()
```
